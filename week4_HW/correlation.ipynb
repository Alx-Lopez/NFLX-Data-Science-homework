```python
# Import libraries and dependencies
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
# Read the CSV file into a `DataFrame` and set the date column as the index
file_path = "../resources/stocks.csv"
stocks_df = pd.read_csv(file_path)
stocks_df
```

```python
# Use the `pct_change` function in Pandas to calculate daily returns for each stock and drop N/A values
stocks_df['JNJ'] = stocks_df['JNJ'].fillna(0).map(lambda x: float(x))
stocks_df["HD"] = stocks_df["HD"].fillna(0).map(lambda x: float((x)))
stocks_df["INTC"]= stocks_df["INTC"].fillna(0).map(lambda x: float((x)))
stocks_df["AMD"] = stocks_df["AMD"].fillna(0).map(lambda x: float((x)))
stocks_df["MU"]= stocks_df["MU"].fillna(0).map(lambda x: float((x)))
stocks_df["NVDA"] = stocks_df["NVDA"].fillna(0).map(lambda x: float((x)))
stocks_df["TSM"]= stocks_df["TSM"].fillna(0).map(lambda x: float((x)))

non_date_df =stocks_df.drop(['date'], axis=1)
non_date_df.pct_change()
```

```python
# Use the `corr` function to calculate correlations of their daily returns for each stock pair
non_date_df.corr()
```

```python
# Create a heatmap from the correlation values
ProfileReport(non_date_df)
```
